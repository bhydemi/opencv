```python
import cv2 
from matplotlib import pyplot as plt
```

```python
ice_flakes = cv2.imread('./images/ice-flakes-microscopy-salt-and-pepper-noise.jpg')
```

```python
plt.imshow(ice_flakes[:,:,::-1])
```

```python
ice_flakes_blurred = cv2.medianBlur(ice_flakes, 3 )
plt.imshow(ice_flakes_blurred[:,:,::-1])
```

```python
face = cv2.imread('images/face-original.jpg')
girl_face = cv2.imread('images/girl-skin.jpg')

smoothen_face = cv2.bilateralFilter(face, 25, 90, 40)
smooth_girl_face = cv2.bilateralFilter(girl_face, 30, 65, 15)

plt.figure(figsize=(12, 6))
plt.subplot(221); plt.imshow(face[:,:,::-1]); plt.axis('off'); plt.title('Original Face')
plt.subplot(222); plt.imshow(girl_face[:,:,::-1]); plt.axis('off'); plt.title('Original Face')
plt.subplot(223); plt.imshow(smoothen_face[:,:,::-1]); plt.axis('off'); plt.title('Smoothen Face')
plt.subplot(224); plt.imshow(smooth_girl_face[:,:,::-1]); plt.axis('off'); plt.title('Smoothen Face')
```
