```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['image.cmap'] = 'gray'
```

```python
forest_lizard = cv2.imread('visuals/Large_Scaled_Forest_Lizard.jpg')
```

```python
forest_lizard_gray = cv2.cvtColor(forest_lizard, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(15, 12))
plt.subplot(121); plt.imshow(forest_lizard[:,:,::-1])
plt.subplot(122); plt.imshow(forest_lizard_gray)
```

```python
kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
```

```python
forest_lizard_gray_m = cv2.filter2D(forest_lizard_gray, kernel=kernel, ddepth=cv2.CV_64F)
forest_lizard_canny = cv2.Canny(forest_lizard_gray, threshold1=180, threshold2=255)
forest_lizard_gray_blurred = cv2.GaussianBlur(forest_lizard_gray, (5,5), 0)
forest_lizard_gray_canny = cv2.Canny(forest_lizard_gray_blurred, threshold1=180, threshold2=255)

plt.figure(figsize=(15, 7.5))

plt.subplot(231); plt.axis('off'); plt.imshow(forest_lizard[:, :,::-1])
plt.subplot(232); plt.axis('off');  plt.imshow(forest_lizard_canny)
plt.subplot(233); plt.axis('off');  plt.imshow(forest_lizard_gray_blurred)
plt.subplot(234); plt.axis('off');  plt.imshow(forest_lizard_gray_canny)
plt.tight_layout()
```
